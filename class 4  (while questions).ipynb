Q1:Write a Python program that asks the user to keep entering numbers until they enter the same number three times in a row. At the end, print:
	1.	The total count of numbers entered.
	2.	The sum of all numbers entered.
	3.	The average of the numbers entered (rounded to 2 decimal places).

```python
count = 0
total = 0
ncount = 0
prevnum = None

while True:
    # Stop if same number entered 3 times in a row
    if ncount == 3:
        break

    try:
        num = int(input("Enter a number: "))
        count += 1
        total += num

        # Check repetition
        if prevnum == num:
            ncount += 1
        else:
            ncount = 1  # Start counting again from 1 for the new number

        prevnum = num

    except ValueError:
        print("Invalid input! Enter again.")
if count > 0:
    average = round(total / count, 2)
    print("\n--- Results ---")
    print("Total count:", count)
    print("Sum:", total)
    print("Average:", average)


    
```

Q:Keep asking for `positive numbers` and add them to a running total until the total ≥ target (ask the target first).


`Print: total, number of entries, and overshoot (total − target)`.


`Note: Ignore non-positive inputs.`

```python
total = 0
entries = 0
target = int(input("enter a target"))
while total < target:
    try:
        num = int(input("enter a number"))
        entries +=1
        if num<0:
            continue
        total = total + num 
    except ValueError:
        print("incorrect data type entered")
print(f"{entries} were the total attempts made including positive and negative entries")
print(f"{total-target} was the overshoot")
```

Ask the user to set a 4-digit PIN (digits only). Then repeatedly ask them to enter the PIN to unlock.
	
    •	If non-digits or wrong length → “Invalid format.”
	
    
    •	Max 3 attempts → lock after that.
Print: “Unlocked” or “Locked”.

```python
'''
pin will initialized as empty string and string function i.e isdigit(), len(), strip() will be used
'''
pin = ""
attempt = 0

#Set a 4-digit PIN
while len(pin) != 4:
    try:
        pin = input("Enter a 4-digit PIN: ")
        pin = pin.strip() #to remove any spaces, hence special characters will be taken as incorrect format
        if not pin.isdigit() or len(pin) != 4:
            print("Wrong length or format, enter again.")
            pin = "" 
    except ValueError:
        print("Incorrect format entered of PIN, please enter again.")

#Unlock attempts
while attempt != 3: #counting attempts
    try:
        pin2 = input("Enter a 4-digit PIN to unlock: ")
        pin2 = pin2.strip()
        attempt += 1

        if not pin2.isdigit() or len(pin2) != 4:
            print("Invalid PIN format.")
            continue

        if pin2 == pin:
            print("Unlock")
            break
        else:
            print("Wrong PIN, try again.")

    except ValueError:
        print("Incorrect format entered.")

#Lock 
if attempt == 3 and pin2 != pin:
    print("Locked")
```

Q:Keep reading integers until user types stop. Ignore inputs that aren’t valid integers.


Print: min, max, and count of valid numbers read.


Edge: If no valid numbers → print “No data”

```python
minimum = None
maximum = None
validcount = 0

while True:
    x = input("Enter an integer (or type STOP to end): ").strip().lower()
    if x == "stop":
        break

    try:
        num = int(x)  
    except ValueError:
        print("invalid input")
        continue
    validcount += 1
    if maximum is None or num > maximum:
        maximum = num
    if minimum is None or num < minimum:
        minimum = num


if validcount == 0:
    print("No data")
else:
    print("Valid count =", validcount)
    print("Min =", minimum)
    print("Max =", maximum)
```
