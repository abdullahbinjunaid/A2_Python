Q2:Write a function table(n) that prints the multiplication table of a number from 1 to 10.
(No return value, only printing.)

```python
def table(num):
    for i in range(1,11):
        print(num,"x",i,"=",num*i)
    return
x = int(input("enter an integer"))
table(x)
```

Q3: Write a function print_evens(limit) that prints all even numbers from 1 to limit.



```python
def limit(num):
    for i in range(1,num+1):
        if i%2 == 0:
            print(i)
    return
x = int(input("enter a number range"))
limit(x)
```

Q4:Write a function box(word) that prints the word inside a box of *.

```python
def box(name):
    y = len(name)
    print(f"#{y*"*"}#")
    print(f"#{name}#")
    print(f"#{y*"*"}#")
    return
x = input("enter a name")
box(x)
    
```

Q5:Write a function report(name, marks) that prints a small report card with name and marks.


```python
def report(name,marks):
    print(f"#student: {name}")
    print(f"#Marks: {marks}")
    return
x = input("enter your name")
y = int(input("enter your marks"))
report(x,y)
```

Q1:Write a function square(num) that returns the square of a number.




```python
def square(num):
    sqr=num**2
    return sqr
x = int(input("enter a number"))
squared = square(x)
print(squared)
```

Q2:Write a function maximum(a, b, c) that returns the largest of the three numbers.


```python
def maximum(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c
a = int(input("enter first number"))
b = int(input("enter secound number"))
c = int(input("enter third number"))
max =  maximum(a,b,c)
print(f"the max is {max}")
```
