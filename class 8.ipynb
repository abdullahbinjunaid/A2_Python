# LINEAR SEARCH

In **python** we know that the `length of array will be known` hence the examiner can first ask us to `make array` and do `some commands or store some data` in it, then `linear search algorithm` will be applied.

*Similar to psuedocode in python there will be 3 cases for searching in linear search*

CASE A:  The element is stored once in the array of known length 

```python
arr = [20,30,40,-50,60,-100]
find = 20
i = 0
while (i<len(arr)) and (arr[i] != find):
    i = i + 1
if i<len(arr):
    print(i,"is the postition of", find)
else:
    print(find,"not found")
```

CASE B: The element is stored more than once, known no. of times, in the array of known length 

```python
arr = [20,30,20,-50,60,-100]
find =  20
amount = 0
i = 0
while (i<len(arr)) and (amount != 2):
    if (arr[i] == find):
        amount = amount + 1
        print(find,"is found at index", i)
    i = i + 1
if amount>0:
    print(amount,"is the amount of times", find,"was found")
else:
    print(find,"was not found")
```

CASE C: The element os stped unkown no. of times in the array, of known lenght,

`NOTE: In python as said before the array length will be known as it will be defined by us`

```python
arr = [20,30,20,-50,60,-100]
find = 20
amount = 0
for i in range(0,len(arr)):
    if arr[i] == find:
        amount += 1
        print(f"{find} was found at {i}")
print(f"{find} was found {amount} number of times")
```

# BINARY SEARCH

array needs to be sorted.

middle value = (lower bound + upper bound) /2

```python
Lb = 0
Ub = len(arr)
found = False
find = 2
while (found = False) and (Lb <= Ub):
    Mv=int((Lb+Ub)/2)

    if (find > arr[Mv]):
        Lb = Mv + 1
    elif (find < arr[Mv]):
        Ub = Mv - 1
    else:
        found = True
if found == True:
    print(f"{Mv} is the position of {find}")
else:
    print(f"{find} not found")
```
