```python
#Question2: slicing a string and using if-condition on it.
text = input("input a text to slice it: ")
if len(text)>10:
    print(text[5:])
else:
    print(text.upper())
```

Question: enter a string, if the position is less than half the length of the string, print the substring from the start to the position entered by the user, else print the string in lower case.


```python
text =  input("input a text")
pos = int(input("input a position for string slicing"))
if pos<(len(text)/2):
    print(text[:pos])
else: 
    print(text.lower())
```

Question: take a character and tell whether its a digit, letter both upper ya lower case, or its a special case character

```python
text = input("input a character: ")
print(text)
text = text.strip("-")
if text.isalpha():
    if text.isupper():
        print("upper case character")
    else:
        print("lower case character")
elif text.isdigit():
    print("Digit")
else:
    print("special character")
    
```

# Case-Of

```python
option = input("enter an option")
match option:
    case "A":
        print("incorrect")
    case "B":
        print("incorrect")
    case "C":
        print("correct")
    case "D":
        print("incorrect")
    case _:
        print("incorrect option entered")
```

```python
value = int(input("enter a number of your choice"))
match value:
    case value if 5<=value<=20:
        print("wow")
    case value if 21<=value<=60:
        print("small")
```
