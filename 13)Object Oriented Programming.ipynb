# OBJECT ORIENTED PROGRAMMING 
___

**CLASS** :  Design/Blueprint. A class is a user-defined data type that defines the attributes and behaviors of objects. It defines:
1. `ATTRIBUTES` : data belonging to the object.
2. `METHODS` : functions belonging to the object.


**OBJECT** : actual item made from the design, or we can say instance of a class.It represents an actual entity created using the class blueprint.



```python
#Example of class and object.
#Class human is made with an attribute height.
class Human:
  def __init__(self,height): # __init__ is a constructor
    self.height = height
  def displayheight(self):
    print(self.height)
#zeek is an object, which defined by class human and has height 5.11
zeek = Human(5.11)
zeek.displayheight()
```

### WHAT IS A CONSTRUCTOR?

A constructor is a special method of a class that is called automatically when an object is created and is used to initialise the object's attributes with starting values.

*In Python* :` __init__` method
___
**NOTE** : A constructor only prepares attributes. It does NOT return anything.

BEFORE constructor is called, object existed but it did'nt had any attribute, but AFTER the constructor was called and it was run, the attributes begin to exists for that object, and these attributes are only accesible by only using the object.
___

CONSTRUCTOR ONLY RUNS ONCE PER OBJECT.

This explains why objects do not overwrite each other's data
___
` If two constructors are created inside a class only the second constructor is run (typically the last one)`

```python
class Human:
  def __init__(self,height): 
    self.height = height
  def displayheight(self):
    print(self.height)


zeek = Human(5.11) # (1) Python converts this internally into :
                   # Human.__init__(zeek,5.11)
zeek.displayheight()
```

(1) Zeek was passed automatically, this is called implicit object passing 

### CLASS VARIABLE VS INSTANCE VARIABLE 
___
`CLASS VARIABLE`: A variable that is shared by all the objects of a class and is defined inside the class, BUT OUTSIDE THE CONSTRUCTOR.

`INSTANCE VARIABLE`: A variable, created inside the constructor using self, that belongs only to a specified object to store data unique to that object 

`**CONSTRUCTOR SHOULD NOT DO WORK - DESIGN PRINCIPLE**`

Constructor should not:
1. print messages
2. input data from the user
3. run logic loops

Constructor should:
1. assign values
2. validate data: This ensures that the object never starts in invalid state and wrong data is either rejected or corrected.

```python
class Human:
  def __init__(self, height=None, size=None):
    self.height = height
    self.size = size
  def displayheight(self):
    print(self.height)

zeek = Human()
zeek.height = 5.11
zeek.size = 40
print(zeek.height)
```
