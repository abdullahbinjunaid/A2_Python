Q3: Write a program using a for loop to calculate the factorial of a given number.


::there are two ways of doing this question one is the conventional method of using for loop other is to use the math module of the python.

```python
#method 1
num = int(input("enter a number to calsulate its factorial"))
fact = 1
if num<0:
    print("-1, negative numbers dont have factorial")
elif num == 0:
    print("1, is the factorial")
else:

    for i in range(1,num+1):
        fact = fact * i
    print(f"{fact} is the factorial of {num}")
```

```python
#method 2
#math.factorial() is not defined for negative integers
import math as mt #importing the math function as mt
num = int(input("enter a number to calculate its factorial"))
if num<0:
    print("-1, negative numbers dont have factorial")
elif num == 0:
    print("1, is the factorial")
else:
    print(mt.factorial(num)) #using the math function
```

Q8: Print all prime numbers between 2 and 50 using for loop 

```python
for num in range(2, 51):
    is_prime = True
    for i in range(2, int(num**0.5) + 1): #or we can go till only num, but sqrt of num is an optimization
        if num % i == 0:
            is_prime = False #the boolean is used to check for a number whether is it prime and break inner for loop for further checking
            break
    if is_prime:
        print(num)
```

Q10:Ask the user to enter 10 integers. Using a for loop, count how many are odd and how many are even.


```python
countE=0
countO=0
for i in range(10):
    num = int(input("enter a number: "))
    print(num)
    if num%2==0:
        countE = countE + 1
    else:
        countO = countO + 1
print(f"{countE} is the amount of even numbers you entered, {countO} is the amount of odd numbers you entered")
```

Q11: Ask the user to enter 5 words. Use a for loop to check each one and print whether it’s a palindrome or not.

```python
nstring=""
for i in range(5):
    nstring=""
    str=input("enter a string to check whether its a palindrome or not")
    str = str.lower()
    for z in range(len(str)-1,-1,-1):
        nstring = nstring + str[z]
    print(nstring)    
    if nstring==str:
        print(f"word {str} ---> palindrome")
    else:
        print(f"word {str} ---> not a palindrome")
```

Q13: Ask the user for a string and use a for loop to count how many times each character appears.

Expected output:

Enter: banana  
b → 1  
a → 3  
n → 2

```python
txt = input("enter a word")
print(txt)
txt = txt.lower()
for i in txt:
    count=0
    for j in range(len(txt)):
        if txt[j]==i:
            count +=1
    print(f"{i} ---> {count}")
```
