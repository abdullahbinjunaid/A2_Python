```python
#method 1: noob 

arr = [2,3,4] 
print(arr)

#method 2
arr = [] #if we want to put a value in this array we use .append eg: arr.append(22)
print(arr)
arr.append(1)
print(arr)

#method 3
#used when we need/ examiner ask to make an empty array 
arr = [0] * 5
print(arr)
arr = [None] * 5
print(arr)

#method 4
import array
arr = array.array('d',[1,2,3,4]) #f:float, d:double, i:signed integer
print(arr)

#method 5
arr = list(range(5))
print(arr)

#method 5
arr = list(range(5))
arr = list("hello")
print(arr)

#method 6
# used when we want to make customized array
arr = [x**2 for x in range (5)]
print(arr)
```

`NOTE`

List in Python is already global, so it isnt neccessary to pass it down into a function or void functions, as changes inside the function will be reflected in the original list whether passed down or not, or if we are doing linear search just use the name of original list in the function and searching will be done

***Example 1:***

```python
arr = [4, 7, 4, 9, 4]   # global list

def linear_search(find):
    first = -1
    count = 0
    for i in range(len(arr)):   # directly using global arr
        if arr[i] == find:
            count += 1
            if first == -1:
                first = i

    if count > 0:
        print(f"Found at index {first}")
        print(f"{find} appeared {count} time(s)")
    else:
        print(f"{find} not found")

linear_search(4) 
```

***Example 2:***

```python
arr = [1, 2, 3]

def change_global_list():
    arr.append(100)       # modifies global list
    arr[1] = 999

change_global_list()
print("After change:", arr)
```

`global list` is to be used when we are assigning an entirly new list to the original value

```python
arr = [1, 2, 3]

def replace_list():
    global arr
    arr = [9, 9, 9]  # completely reassigning the global variable

replace_list()
print(arr)
```

But the above practice (not global one) isnt optimal because array can varry and different list can be used with the same function hence, we would take the approach of `sending the list as parameter.`

This approach will be best as the same function can be used for different list if doing linear search, sorting, binary search and etc



***Example Codes are below***

```python
def linear_search(arr, find):
    first = -1
    count = 0
    for i in range(len(arr)):
        if arr[i] == find:
            count += 1
            if first == -1:
                first = i

    if count > 0:
        print(f"Found at index {first}")
        print(f"{find} appeared {count} time(s)")
    else:
        print(f"{find} not found")

numbers = [4, 7, 4, 9, 4]
linear_search(numbers, 4)
```
