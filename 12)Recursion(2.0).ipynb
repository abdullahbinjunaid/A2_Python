### Implementing Binary Search while using recursion

```python
# Binary search code (without using rescursion)
arr = [3,4,6,8,10,11,15]
Lb = 0
Ub = len(arr)
found = False
find = 11
while (found == False) and (Lb <= Ub):
    Mv=int((Lb+Ub)/2)

    if (find > arr[Mv]):
        Lb = Mv + 1
    elif (find < arr[Mv]):
        Ub = Mv - 1
    else:
        found = True
if found == True:
    print(f"{Mv} is the position of {find}")
else:
    print(f"{find} not found")
```

```python
# Binar search with recursion 
def binarysearch(find,Lb,Ub):

  if Lb > Ub:
    return False
  Mv=int((Lb+Ub)/2)

  if (find > arr[Mv]):
    return binarysearch(find,Mv+1,Ub)
  elif (find < arr[Mv]):
    return binarysearch(find,Lb,Mv-1)
  else:
    print(f"{find} was found at {Mv}")
    return True


arr = [3,4,6,8,10,11,15]
Lb = 0
Ub = len(arr) -1
find = 19

print(binarysearch(find,Lb,Ub))

```
