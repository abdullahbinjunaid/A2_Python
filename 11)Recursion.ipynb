# RECURSION
___
A process where a function calls itself or indirectly to solve a problem. it breaks the problem into simpler subproblems till the base case arise.

`2 components of recursion:`
---
1) **BASE CASE**: The condition where the recursion stop. *without a base case, an infinite recursion will occur giving  a recursion error
2) **RECURSIVE CASE**: The part of the funtion where it calls itself to handle smaller instances of the problem.



***HOW IT WORKS*** 
___
Each function call is added to the call stack.

when a function finishes, it is removed from the stack and the program returns to the previous call

`CALL STACK`:The call stack is a fundamental data structure that Python uses to manage function calls, including recursive ones. It operates on a Last-In, First-Out (LIFO) principle, much like a stack of plates: the last function called is the first one to finish and be removed from the stack. 

### There are two types of Recursion.

1) `Direct`: A function calls itself
2) `Indirect`: A function calls another function which in return calls the first function back .

```python
def factorial(n):
  # Base Case: if n is 0 or 1, return 1
  if n== 0 or n==1
    return 1
  # Recursive case: n * factorial(n-1)
  else:
    return n * factorial(n-1)

number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")
```
