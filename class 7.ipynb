# ARRAY IN PYTHON

Array in python is called list()...

```python
#method 1: noob 

arr = [2,3,4] 
print(arr)

#method 2
arr = [] #if we want to put a value in this array we use .append eg: arr.append(22)
print(arr)
arr.append(1)
print(arr)

#method 3
#used when we need/ examiner ask to make an empty array 
arr = [0] * 5
print(arr)
arr = [None] * 5
print(arr)

#method 4
import array
arr = array.array('d',[1,2,3,4]) #f:float, d:double, i:signed integer
print(arr)

#method 5
arr = list(range(5))
print(arr)

#method 5
arr = list(range(5))
arr = list("hello")
print(arr)

#method 6
# used when we want to make customized array
arr = [x**2 for x in range (5)]
print(arr)
```
