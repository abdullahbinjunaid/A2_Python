Q:Write a Python program that asks the user to keep entering numbers until they enter the same number three times in a row. At the end, print:
	1.	The total count of numbers entered.
	2.	The sum of all numbers entered.
	3.	The average of the numbers entered (rounded to 2 decimal places).
___

```python
same = 0
numcount = 0
sum = 0 
num = int(input("enter a number"))
prevnum = num 
same +=1
sum = sum + num
numcount +=1
while same != 3:
    try:
        num = int(input("enter a number"))
        numcount += 1
        sum =  sum + num 
        if num == prevnum:
            same += 1
        else:
            same = 1
            prevnum = num 
    except ValueError:
        print("incorrect input")
print(sum)
print(round(sum/numcount),2)
    
```

Q2:Write a Python program that keeps asking the user to guess a secret number between 1 and 20.
	•	If the guess is too high, print "Too high! Try again."
	•	If the guess is too low, print "Too low! Try again."
	•	Stop the loop only when the correct number is guessed.
	•	At the end, print how many attempts the user took.
___    

```python
import random 
guess = random.randint(1,20)
attempts = 0
num = None
while num != guess:
    num = int(input("guess the secret number b/w 1-20"))
    attempts+=1
    if num>guess:
        print("too high")
    if num<guess:
        print("too low")
    if num ==  guess:
        print(f"correct: you found {guess} in {attempts} attempts")
```

	1.	The largest number entered.

Important: Ignore -1 in calculations.
___
