# STACKS

LIFO : Last in First Out
____
Stack Works with this concept Last in First Out, A single variable Top is used to push or pop a data in the stack, PUSH is adding a data value inside the stack in which the TOP value is incremented by 1 and then the data value is added, While the POP is done to remove the data from the stack, its done by decrementing the TOP value by 1, but first remove the data from the stack and place it inside a variable before decrementing the TOP Value



```python
# if top == len(stack) - 1 (condition to check whether the stack is full or not)
# top == -1 (stack is empty condition)
# PUSH : Top = Top + 1
#        Top = value(data)
```

```python
#PUSH command for stack
def push(stack):
  global Top
  if Top == len(stack) - 1:
    print("the stack is full")
  else:
    data = int(input("enter a digit you want to put in stack"))
    Top = Top + 1
    stack[Top] = data
    print(stack)
  

#POP command for stack 
def pop(stack):
  global Top
  if Top == -1:
    print("stack is empty")
  else:
    value = stack[Top]
    stack[Top] = None
    Top = Top - 1
    print(f"The value which was popped was {value}")
    print(stack)


stack = [None]*10
Top = -1
choice = 0
try:
  while (choice != -1):
    choice = int(input("enter 1 for push, 2 for pop, -1 to end the program"))
    if choice == 1:
      push(stack)
    elif choice == 2:
      pop(stack)
    else:
      print("program ending")
except ValueError:
  print("incorrect value entered")
else:
  print(stack)
finally:
  print("program completed")
```

## LINEAR QUEUE

linear queue has two pointers `END` and `FRONT`. As the `data is added from the END`, it `moves forward` as new data values are being inputted, while the `FRONT is used to remove data` from the `starting of the queue`. `Inputting a data is known as ENQUEUE` and `Removing a data is called DEQUEUE`. Initially both front and end pointer have the `index of -1` as this indicates the queue is empty. When first data is entered, its important to increment both front and `end by +1`, this has to be done for the first input, after that when adding a `new input` the end is only to be incremented and front is to be inceremented when a data value is to be removed.
