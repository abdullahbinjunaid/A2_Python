
# STACK

LIFO : Last in First Out




```python
# if top == len(stack) - 1 (condition to check whether the stack is full or not)
# top == -1 (stack is empty condition)
# PUSH : Top = Top + 1
#        Top = value(data)
```

```python
def push(data,stack):
  global top

  top = top + 1
  stack[top] = data
def pop(stack):
  global top
  value = stack[top]
  stack[top] = None
  top = top - 1
  return value


stack = [None]*10
top = -1
choice = 0
try:
  while (choice != -1):
    choice = int(input("enter 1 for push, 2 for pop, -1 to end the program"))
    if choice == 1:
      if top == len(stack) - 1:
        print("the stack is full")
      else:
        data = int(input("enter a data value to push in stack"))
        push(data,stack)
        print(stack)
    elif choice == 2:
      if top == -1:
        print("stack is empty")
      else:
        value = pop(stack)
        print(f"the value which was popped was {value}")
        print(stack)

    else:
      print("program ending")
except ValueError:
  print("incorrect value entered")
else:
  print(stack)
finally:
  print("program completed")




```

# QUEUE

FIFO : First in First Out

```python
# 2 pointers: End, Front
# Data In : Enqueue
# Data Out : Dequeue
# element is added from the End
# if End == len(Queue) - 1 (condition for queue to be full)
''' if front == end + 1:
      print("queue is empty")
      End = -1
      Front = -1 (this is the condition to check if the queue is empty)
'''
```

```python
def Enqueue(data,queue):
  global front
  global end
  if end == len(queue) - 1:
    print("queue is full")
  else:
    if front == -1:
      front = front + 1
    end = end + 1
    queue[end] = data
  print (front,end)

def Dequeue(queue):
  global front
  global end
  if front == end + 1:
      print("queue is empty")
      end = -1
      front = -1
  elif (front != -1):
    value = queue[front]
    queue[front] = None
    front = front + 1
    print(f"the value which was removed was {value}")
  print(front,end)

queue = [None]*5
front = -1
end = -1
choice = 0
try:
  while (choice != -1):
    choice = int(input("enter 1 for enqueue, 2 for dequeue, -1 for terminating the program"))
    if choice == 1:
      data = int(input("enter a data to enqueue"))
      Enqueue(data,queue)
      print(queue)
    elif choice == 2:
      Dequeue(queue)
      print(queue)
    else:
      print("invalid input")
except ValueError:
  print("incorrect data type entered")
else:
  print(queue)
finally:
  print("program completed")
```
