# BUBBLE SORT

Bubble sorting approach will be same as the approach we did in psuedocode, this was done by setting a swap variable false and going from the first element of the array till Top-1 element of the array and compare the index element with index+1 element of the array and swap accordingly, decending or ascending. After swapping the swap varaible will be set to true. Continue this unless the Top variable becomes 1 or the swap varaible remains False-meaning no swapping was done the array is sorted.

```python
# In python it will be done like this 
arr = [-10,-20,5,2,3,10]
Top = len(arr)-1
swap = True
while (Top>1) and (swap == True):
    swap = False
    for i in range(Top):
        if ( arr[i] > arr[i+1] ):
            temp = arr[i]
            arr[i] = arr[i+1]
            arr[i+1] = temp
            swap = True
    Top = Top - 1

print(arr)
```
